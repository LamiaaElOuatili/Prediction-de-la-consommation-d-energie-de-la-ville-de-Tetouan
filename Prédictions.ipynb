```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv(r"C:\Users\jh\Desktop\Comp\train_data.csv")
df
```

```python


feature_columns = ["Temperature","Humidity","WindSpeed","GeneralDiffuseFlows","DiffuseFlows"]
X = df[feature_columns]
X
```

```python
feature_columns1 = ["PowerConsumption_Zone1"]
y1 =  df[feature_columns1]

feature_columns2 = ["PowerConsumption_Zone2"]
y2 =  df[feature_columns2]

feature_columns3 = ["PowerConsumption_Zone3"]
y3 =  df[feature_columns3]
```

```python

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
def get_mae1(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python

# fill in and uncomment
train_X, val_X, train_y, val_y = train_test_split(X, y2, random_state=1)


candidate_max_leaf_nodes = [16500, 17500, 20000, 21000]
# Write loop to find the ideal tree size from candidate_max_leaf_nodes

L = [get_mae(i, train_X, val_X, train_y, val_y) for i in candidate_max_leaf_nodes]

# Store the best value of max_leaf_nodes (it will be either 5, 25, 50, 100, 250 or 500)
best_tree_size = L.index(min(L))
best_tree_size
```

```python
# fill in and uncomment
train_X, val_X, train_y, val_y = train_test_split(X, y3, random_state=1)


candidate_max_leaf_nodes = [1000, 1100, 1400, 1500, 1550]
# Write loop to find the ideal tree size from candidate_max_leaf_nodes

L = [get_mae1(i, train_X, val_X, train_y, val_y) for i in candidate_max_leaf_nodes]

# Store the best value of max_leaf_nodes (it will be either 5, 25, 50, 100, 250 or 500)
best_tree_size = min(L)
print(best_tree_size)
print(L.index(min(L)))
```

```python
iowa_model1 = RandomForestRegressor(max_leaf_nodes=17000, random_state=1)
iowa_model1.fit(X, y1)
pred1 = iowa_model1.predict(X)
val_mae = mean_absolute_error(pred1, y1)

# uncomment following line to see the validation_mae
print(val_mae)
```

```python
iowa_model2 = DecisionTreeRegressor(max_leaf_nodes=1200, random_state=1)
iowa_model2.fit(X, y2)
pred2 = iowa_model2.predict(X)
val_mae = mean_absolute_error(pred1, y2)

# uncomment following line to see the validation_mae
print(val_mae)
```

```python
iowa_model3 = DecisionTreeRegressor(max_leaf_nodes=12000, random_state=1)
iowa_model3.fit(X, y3)
pred3 = iowa_model3.predict(X)
val_mae = mean_absolute_error(pred1, y3)

# uncomment following line to see the validation_mae
print(val_mae)
```
